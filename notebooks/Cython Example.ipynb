```python
import array
a = array.array('l',range(100))
s = 0
```

Sum up an array of numbers using python

```python
def python_sum(a):
    global s
    s = 0
    for i in range(len(a)):
        for j in range(10000):
            s = s + a[i]
    return s
```

```python
%timeit python_sum(a)
```

Use cython, without changing the code

```python
%%cython --annotate
def cython_sum1(a):
    global s
    s = 0
    for i in range(len(a)):
        for j in range(10000):
            s = s + a[i]
    return s
```

```python
print('python sum: ',python_sum(a))
print('cython sum1: ',cython_sum1(a))
print('python sum')
%timeit python_sum(a)
print('cython sum1')
%timeit cython_sum1(a)
```

Does making s a local variable help?

```python
%%cython --annotate
def cython_sum2(a):
    s = 0
    for i in range(len(a)):
        for j in range(10000):
            s = s + a[i]
    return s
```

```python
print('python sum: ',python_sum(a))
print('cython sum1: ',cython_sum1(a))
print('cython sum2: ',cython_sum2(a))
print('python sum')
%timeit python_sum(a)
print('cython sum1')
%timeit cython_sum1(a)
print('cython sum2')
%timeit cython_sum2(a)
```

```python
%%cython --annotate
from cpython cimport array

def cython_sum3(a):
    cdef long s = 0
    cdef array.array ta = a
    cdef long * ap = ta.data.as_longs
    for i in range(len(ta)):
        for j in range(10000):
            s = s + ap[i]
    return s
```

```python
print('python sum: ',python_sum(a))
print('cython sum1: ',cython_sum1(a))
print('cython sum2: ',cython_sum2(a))
print('cython sum3: ',cython_sum3(a))
print('python sum')
%timeit python_sum(a)
print('cython sum1')
%timeit cython_sum1(a)
print('cython sum2')
%timeit cython_sum2(a)
print('cython sum3')
%timeit cython_sum3(a)
```

```python
from numba import jit
@jit
def numba_sum(a):
    s = 0
    for i in range(len(a)):
        for j in range(10000):
            s = s + a[i]
    return s
```

```python
print('python sum: ',python_sum(a))
print('cython sum1: ',cython_sum1(a))
print('cython sum2: ',cython_sum2(a))
print('cython sum3: ',cython_sum3(a))
print('numba sum: ', numba_sum(a))
print('python sum')
%timeit python_sum(a)
print('cython sum1')
%timeit cython_sum1(a)
print('cython sum2')
%timeit cython_sum2(a)
print('cython sum3')
%timeit cython_sum3(a)
print('numba sum')
%timeit numba_sum(a)
```
